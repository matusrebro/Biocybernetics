```python
from Diabetes import minimal_model # importing the minimal model
```

```python
# glycemic response vs. glycemic index 
# initialize model for normal subject
Gb = 5 # [mmol/L]
Ib = 6 # [mU/L]
model_normal_oral = minimal_model.oral(Gb, Ib, parameters = 'normal')
# initialize model for t2dm subject
Gb = 7 # [mmol/L]
Ib = 7 # [mU/L]
model_t2dm_oral = minimal_model.oral(Gb, Ib, parameters = 't2dm')
```

```python
# simulation for normal subject - high glycemic index
glucose = 50 # total amount of carbohydrates in food
BW = 70 # bodyweight of a subject
G, I = model_normal_oral.ogtt(glucose, BW, 'high')
```

```python
# simulation for normal subject - low glycemic index
G, I = model_normal_oral.ogtt(glucose, BW, 'low')
```

```python
# simulation for t2dm subject - high glycemic index
glucose = 50 # total amount of carbohydrates in food
BW = 100 # bodyweight of a subject
G, I = model_t2dm_oral.ogtt(glucose, BW, 'high')
```

```python
# simulation for normal subject - low glycemic index
G, I = model_t2dm_oral.ogtt(glucose, BW, 'low')
```
