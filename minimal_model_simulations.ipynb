```python
from Diabetes import minimal_model # importing the minimal model
```

```python
# initialize model for intravenous input
Gb = 5 # [mmol/L]
Ib = 6 # [mU/L]

model_norma1 = minimal_model.iv(Gb, Ib, parameters = 'normal')
```

```python
# IVGTT simulation for normal subject
glucose_dose = 0.3 # g/kg
glucose_bolus_min = 2 # min - glucose bolus is administred over 2 minutes
insulin_dose = 20 # mU/kg
insulin_bolus_min = 5 # min - insulin bolus is administred over 5 minutes
insulin_dosage_time = 20 # min - insulin administration starting at 20 minutes from the start of the test
G, I  = model_norma1.ivgtt(glucose_dose, glucose_bolus_min, insulin_dose, insulin_bolus_min, insulin_dosage_time)
```

```python
# euglycemic glucose clamp simulation for normal subject
BW = 70 # [kg] - bodyweight
G, I, RaG_iv = model_norma1.hyperinsulinemic_euglycemic_glucose_clamp(BW)
```
